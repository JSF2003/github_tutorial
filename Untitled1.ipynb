```python
df = pd.read_html('https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/')
df[0]
```

```python
df=df[0]
```

```python
# to change name of a column
new_names = ['NameBank', 'City', 'State' , 'Cert', 'AI', 'ClosingDate', 'Fund']
```

```python
df.columns = new_names
```

```python
df['date'] = pd.to_datetime(df['ClosingDate'])
```

```python
# dropping the close_date column
df.drop('ClosingDate',axis=1,inplace=True)
df.head()
```

```python
df['date'].dtypes
```

```python
#df['date'] = df['date'].apply(lambda x: x.split('-'))
```

```python
df['date'] = df['date'].astype('string')
```

```python
df['date'].dtypes
```

```python
df['Year']=df['date'].apply(lambda x: x.split('-')[0])
df['Month']=df['date'].apply(lambda x: x.split('-')[1])
df['Day']=df['date'].apply(lambda x: x.split('-')[2])
```
