```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,palette='coolwarm')
```

```python
sns.boxplot(data=tips,palette='Set1',orient='h')
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker',data=tips, palette='coolwarm')
```

```python
sns.boxplot(x='day', y='tip', hue='smoker',data=tips, palette='coolwarm')
```

```python
tips['percent'] = (tips['tip']/tips['total_bill'])
```

```python
sns.boxplot(x='day',y='percent', hue='sex',data=tips, palette='coolwarm')
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker',data=tips, palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex',data=tips, palette='Set1')
```

```python
sns.violinplot(x='day', y='percent', hue='sex',data=tips, palette='Set1')
```

```python
sns.violinplot(x='day', y='percent', hue='sex',data=tips, split=True, palette='rainbow')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill',data=tips,palette='Set1', split=True)
```

```python
sns.swarmplot(x='tip', y='day',data=tips,color='black', size=3)
sns.violinplot(x='tip', y='day',data=tips, palette='rainbow')
```

```python
sns.swarmplot(x='day', y='tip',data=tips,color='black', size=3)
sns.violinplot(x='day', y='tip',data=tips, palette='rainbow')
```
