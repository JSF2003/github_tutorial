```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,5,200)
y = x**2
```

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.3,0.3,0.4,0.5])
ax1.plot(x,y,'r')
ax2.plot(y,x,'g')
```

```python
# subplots
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,6))
axes[0].plot(x,x**2,x,np.exp(x))
axes[0].set_title('Normal Scale of Y')
axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('log scale(Y)')
```

```python
x = np.linspace(0,16,100)
y= x**2
z = np.sqrt(x)
```

```python
fig, ax1 = plt.subplots()
color1 = 'r'
ax1.plot(x,y,color=color1)
ax1.tick_params(axis='y', color=color1)
color2 = 'b'
ax2 = ax1.twinx()
ax2.plot(x,z,color=color2)
ax2.tick_params(axis='y', color=color2)
ax2.spines['left'].set_color(color1)
ax1.spines['right'].set_color(color2)
```
